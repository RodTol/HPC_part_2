```python
import tabulate
import pandas as pd
import numpy as np
```

```python
files = ['times1.dat', 'times2.dat', 'times4.dat', 'times8.dat']
```

```python

lines = []
data_frames = []
n_line = 0
for index in range(len(files)):
	with open(files[index]) as file:	
		for line in file.readlines():
			lines += [line.split()]
			n_line+=1

	for i in range(len(lines[0])):
		lines[0][i] = lines[0][i].replace(",","")
	lines[0].insert(0,"Compilation")

	for line in lines[1:]:
		line[1:] = (np.array(line[1:])).astype(np.double) 

	#lines =  lines[0:len(lines)-3]
	df = pd.DataFrame(lines[1:])
	df.columns = lines[0]
	data_frames.append(df)
	lines = []

for i in data_frames:
	display(i)
```

```python
import matplotlib.pyplot as plt

def calc_comm_time(df):
    comm_time_1 = []
    comm_time_2 = []
    comm_time_3 = []
    for i in range(int(df["comm_total"].shape[0]/3)):
        comm_time_1.append(df["comm_total"][i*3])
        comm_time_2.append(df["comm_total"][i*3+1])
        comm_time_3.append(df["comm_total"][i*3+2])

    comm_time = [comm_time_1, comm_time_2, comm_time_3]
    return comm_time

def calc_compute_time(df):
    compute_time_1 = []
    compute_time_2 = []
    compute_time_3 = []
    for i in range(int(df["compute_total"].shape[0]/3)):
        compute_time_1.append(df["compute_total"][i*3])
        compute_time_2.append(df["compute_total"][i*3+1])
        compute_time_3.append(df["compute_total"][i*3+2])

    compute_time = [compute_time_1, compute_time_2, compute_time_3]
    return compute_time

def calc_name(df):
    name = []
    for i in range(int(df["N"].shape[0]/3)):
        name.append(df["N"][i*3])
    return name
```

```python
def plot(df):
    comm_time = calc_comm_time(df)
    compute_time = calc_compute_time(df)
    name = calc_name(df)
    X = np.arange(len(comm_time[0]))

    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    #ax.bar(X + 0.00, comm_time[0], color = 'b', width = 0.25)
    #ax.bar(X + 0.00, compute_time[0], bottom=comm_time[0], color = 'c', width = 0.25)
    ax.bar(X + 0.25, comm_time[1], color = 'g', width = 0.25)
    ax.bar(X + 0.25, compute_time[1], bottom=comm_time[1], color = 'LightGreen', width = 0.25)
    ax.bar(X + 0.50, comm_time[2], color = 'r', width = 0.25)
    ax.bar(X + 0.50, compute_time[2], bottom=comm_time[2], color = 'm', width = 0.25)

    ticks = ['N = ' +str(size) for size in name]
    ax.set_xticks(X+0.25, (ticks))
    ax.legend(labels=['openBLAS comunication', 'openBLAS computation', 'cuBLAS communication',  'cuBLAS computation'])

    plt.show()
```

```python
for i in data_frames:
    plot(i)
```
